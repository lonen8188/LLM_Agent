```python
# 확장자가 ipynb로 생성하면 자동으로 노트북 버전으로 진행 된다. 오른쪽 상단에 가상환경을 설정 한다. venv로 설정하고 실행버튼을 누르면 하단에 install이 나옴
# install 진행 후 다음 부터 실행 가능
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')
client = OpenAI(api_key=api_key)
```

```python
audio_file_path = '../audio/lsy_audio_2023_58s.mp3' # MP3 파일 경로 입력 (저자의 50초 음성)

# 오디오 파일을 열어서 음성 분석을 진행한다. 
# rb = read binary (바이너리 읽기 모드) , r은 텍스트 모드로 오류남
# wb = white binary (바이너리 쓰기 모드)
with open(audio_file_path, 'rb') as audio_file:
    transcription = client.audio.transcriptions.create(
        # transcription : 음성(말소리) → 텍스트(문자)로 변환한 것
        # translations : 음성 → 다른 언어로 번역
        model="whisper-1", # 위스퍼 모델을 활용 음성 → 텍스트 변환(STT, Speech-to-Text) 할 때 사용하는 모델 이름
        #OpenAI의 음성 인식 모델
        # 오디오 파일(mp3, wav 등)을 텍스트로 변환
        # 다국어 지원 (한국어 포함)
        # model="gpt-4o-mini-transcribe" 최근 많이 사용하는 언어 텍스트 변환 모델
        file=audio_file
    )

transcription
```
